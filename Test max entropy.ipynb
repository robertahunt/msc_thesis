```python
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_clumps = 10
intra_clump_sigma_range = (2,15.5)


mean_range = [0,100]

bins = 100

samples = 10000

clump_params = []
for clump in range(n_clumps):
    mean = mean_range[0] + np.random.random(2)*(mean_range[1] - mean_range[0])
    std = intra_clump_sigma_range[0] + np.random.random(2)*(intra_clump_sigma_range[1] - intra_clump_sigma_range[0])
    clump_params += [np.append(mean,std)]
    
points = []
for i in range(samples):
    params = clump_params[np.random.choice(len(clump_params))]
    mx, my, sx, sy = params
    x = np.random.normal(loc=mx,scale=sx)
    y = np.random.normal(loc=my,scale=sy)
    points += [[x,y]]
points = np.array(points)

N, _, _ = np.histogram2d(points[:,0],points[:,1],bins=10)

N = N.flatten()/samples
H = 0
for p in N:
    if p != 0:
        H -= p*np.log2(p)

plt.scatter(points[:,0],points[:,1],alpha=0.5)
print(H)
```
